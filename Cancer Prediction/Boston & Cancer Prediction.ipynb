```python
import pandas as pd
from sklearn.datasets import load_boston
import numpy as np
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data,columns=boston.feature_names)
df
```

```python
df['target'] = boston.target
```

```python
df.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(boston.data,boston.target,test_size=0.20,random_state=10)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
np.average(cross_val_score(LinearRegression(), cancer.data, cancer.target))
```
